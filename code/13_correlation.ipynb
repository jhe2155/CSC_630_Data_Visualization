```python
import altair as alt
import pandas as pd

# galton_data = pd.io.stata.read_stata('data/galton.dta')
# galton_data.to_csv('data/galton.csv')

df = pd.read_csv('data/galton.csv')

father_chart = alt.Chart(df).mark_point().encode(
    x = alt.X('father', scale=alt.Scale(domain=[min(df['father'])-1, max(df['father']+1)]), axis=alt.Axis(title='Height of Father')),
    y = alt.Y('height', scale=alt.Scale(domain=[min(df['height'])-1, max(df['height'])+1]), axis=alt.Axis(title='Height of Child')),
).properties(
    title="Galton's Correlation Between Height of Father and Child"
)

mother_chart = alt.Chart(df).mark_point().encode(
    x = alt.X('mother', scale=alt.Scale(domain=[min(df['mother'])-1, max(df['mother']+1)]), axis=alt.Axis(title='Height of Mother')),
    y = alt.Y('height', scale=alt.Scale(domain=[min(df['height'])-1, max(df['height'])+1]), axis=alt.Axis(title='Height of Child')),
).properties(
    title="Galton's Correlation Between Height of Mother and Child"
)

father_chart | mother_chart
```
