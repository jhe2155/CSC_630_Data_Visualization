```python
import altair as alt
import pandas as pd

# df = pd.read_csv('data/city_temperature.csv')
# df = df[df['City'] == "Boston"]

# df.to_csv('data/boston_temperature.csv')

df = pd.read_csv('data/boston_temperature.csv')

# df = df[df['Year'] > 2006]

# df.to_csv('data/boston_temperature.csv')

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df = df[df['AvgTemperature'] != -99.0]

df['Month'] = df['Month'].apply(lambda x: months[x - 1])

stripplot =  alt.Chart(df, width=50).mark_circle(size=8).encode(
    x=alt.X(
        'jitter:Q',
        title=None,
        axis=alt.Axis(values=[0], ticks=True, grid=False, labels=False),
        scale=alt.Scale(),
    ),
    y=alt.Y('AvgTemperature', axis=alt.Axis(title='Average Temperature (Fahrenheit)')),
    column=alt.Column(
        'Month',
        header=alt.Header(
            labelAngle=-90,
            titleOrient='bottom',
            labelOrient='bottom',
            labelAlign='right',
            labelPadding=3,
        ),
        sort=months
    ),
).transform_calculate(
    # Generate Gaussian jitter with a Box-Muller transform
    jitter='sqrt(-2*log(random()))*cos(2*PI*random())'
).configure_facet(
    spacing=0
).configure_view(
    stroke=None
).properties(
    title='Average Temperature in Boston from 2007 to 2020, Broken Down by Month'
)

stripplot
```
